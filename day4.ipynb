# Nested for

```python
for i in range(5):
    for j in range(i):
        print('*',end=" ")
    print()
print("Star Printed")    
```

```python
# jump Statements 
# break      escapes current loop
# continue   skips the current iteratiion and iterates to next
# pass       does noting 
```

```python
for i  in range(5):
    if(i==3):
        print('gonna break')
        break
    else:
        print(i)
```

```python
for i  in range(5):
    if(i==3):
        print('gonna continue')
        continue
        print('after continue')
    else:
        print(i)
```

```python
for i  in range(5):
    if(i==3):
        print('gonna pass')
        pass
        print('after pass')
    else:
        print(i)
```
